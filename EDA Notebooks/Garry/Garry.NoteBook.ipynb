```python
plt.figure(figsize=(22, 20))
sns.barplot(data= data_cleaned, x='classification' , y = 'attendance' ,estimator= 'mean', errorbar = None, color= 'red')
plt.title(' Average Attendance  Per Classification')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.barplot(data=data_cleaned, x='borough' , y = 'attendance', hue = 'event_type',estimator= 'mean', errorbar= None )
plt.title('Average Attendance Per Borough w/ Event Type')
plt.xlabel('Borough')
plt.ylabel('attendance')
plt.xticks(rotation=45)
plt.show()


```

```python
# What are the most popular events within each Borough? 

"""
 So as you see the graph above Tournament/Competition event types are the most popular in Queens, Staten Island and
 the Bronx. Staten Island has a big outlier for competition as there was a track and field event, where over 2000 people attended. 
 But overall looking at all the boroughs outide of manhattan, Tournament/Competiton seems to be the most popular 
 event. Manhattan and Brooklyn enjoyed community based events.  The least popular event in every borough (except queens) was open house events. 

"""
```
